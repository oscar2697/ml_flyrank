### **The Research Question**

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

We load the starter anonymized dataset to understand our baseline and verify the volume of opportunities for our Lane 1 study[cite: 3].

```python
import pandas as pd
import numpy as np

# Load the starter dataset from the correct relative path
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# 1. Total initial records
total_records = len(df)

# 2. Filter data as per starter pipeline guidelines (impressions > 0 and age >= 90 days)[cite: 3]
df_clean = df[(df['impressions_90d'] > 0) & (df['content_age_days'] >= 90)].copy()
valid_records = len(df_clean)

# 3. Compute decline rate (our proxy label)[cite: 3]
down_trend_count = (df_clean['trend_direction'] == 'down').sum()
decline_percentage = (down_trend_count / valid_records) * 100

print(f"1. Total Raw Records: {total_records:,}")
print(f"2. Valid Records for Analysis: {valid_records:,} ({(valid_records/total_records)*100:.1f}%)")
print(f"3. Pages currently in decline (trend_direction == 'down'): {down_trend_count:,} ({decline_percentage:.2f}%)")
```
